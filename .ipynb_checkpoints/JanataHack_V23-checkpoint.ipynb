```python
train['Source'] = "Train"
test['Source'] = "Test"

data = pd.concat([train,test],axis=0).reset_index(drop=True)
data.shape
```

```python
data.head()
```

```python
data.info()
```

```python
def get_category(x):
    s = set()
    for item in x.split(";"):
        s.add(item.split("/")[0])
    return list(s),len(list(s))
data['Categories_Viewed'],data['Number_of_Categories_viewed'] = zip(*data['ProductList'].apply(lambda x: get_category(x)))
```

```python
def get_sub_category(x):
    s = set()
    for item in x.split(";"):
        s.add(item.split("/")[1])
    return list(s),len(s)
data['Sub_Categories_Viewed'],data['Number_of_sub_categories_viewed'] = zip(*data['ProductList'].apply(lambda x: get_sub_category(x)))
```

```python
def get_sub_sub_category(x):
    s = set()
    for item in x.split(";"):
        s.add(item.split("/")[2])
    return list(s),len(s)
data['Sub_Sub_Categories_Viewed'],data['Number_of_sub_sub_categories_viewed'] = zip(*data['ProductList'].apply(lambda x: get_sub_sub_category(x)))
```

```python
def get_products(x):
    s = set()
    for item in x.split(";"):
        s.add(item.split("/")[3])
    return list(s)
data['Products_Viewed']= data['ProductList'].apply(lambda x: get_products(x))
```

```python
def create_datetime_cols(df,col):
    words = col[:-4]
    df[words+"_Date"] = df[col].dt.date
    df[words+"_Year"] = df[col].dt.year
    df[words+"_Month"] = df[col].dt.month
    df[words+"_Quarter"] = df[col].dt.quarter
    df[words+"_Weekofyear"] = df[col].dt.weekofyear
    df[words+"_Day"] = df[col].dt.dayofweek
    df[words+"_Weekend"] = df[words+"_Day"].apply(lambda x: 0 if x<=4 else 1)
    df[words+"_is_month_start"] = df[col].dt.is_month_start
    df[words+"_is_month_end"] = df[col].dt.is_month_end
    
    df[words+"_Time"] = df[col].dt.time
    df[words+"_Hour"] = df[col].dt.hour
    df[words+"_Minute"] = df[col].dt.minute
    
    return df
data = create_datetime_cols(data,'startTime')
data = create_datetime_cols(data,'endTime')

data['Number_Of_Products_Viewed'] = data['ProductList'].str.split(";").apply(lambda x: len(x))


data['Time_Difference'] = data['endTime'].subtract(data['startTime'])
data['Time_Difference_In_Seconds'] = data['Time_Difference'].dt.seconds
data['Time_Difference_In_Minutes'] = data['Time_Difference_In_Seconds']/60    

data['Time_Per_Product'] = data['Time_Difference_In_Seconds']/data['Number_Of_Products_Viewed']
```

```python
df = pd.concat([pd.DataFrame(data.Categories_Viewed.values.tolist()).add_prefix('Category_').fillna("Missing"),
           pd.DataFrame(data.Sub_Categories_Viewed.values.tolist()).add_prefix('Sub_Category_').fillna("Missing"), 
           pd.DataFrame(data.Sub_Sub_Categories_Viewed.values.tolist()).add_prefix('Sub_Sub_Category_').fillna("Missing"),
           pd.DataFrame(data.Products_Viewed.values.tolist()).add_prefix('Products_').fillna("Missing")], axis=1)
df.head()
```

```python
df.shape
```

```python
data = pd.concat([data,df],axis=1)
```

```python
data.drop(['endTime','session_id','startTime','start_Date','start_Time','end_Date','end_Time','Time_Difference'],axis=1,inplace=True)
```

```python
cat_cols = [col for col in data.columns if data[col].dtype == "object" if col not in ['gender','Source','Categories_Viewed',
                                                                                     'Sub_Categories_Viewed','Sub_Sub_Categories_Viewed',
                                                                                     'Products_Viewed','ProductList']]
```

```python
cat_cols
```

```python
# Getting Dummy Variables
data = pd.get_dummies(data=data, columns=cat_cols)
```

```python
data.shape
```

```python
# Following columns are dropped from the two datasets as one value dominates the entire column and there wouldn't be much variation..
overfit = []
for col in data.columns:
    if col not in ['Categories_Viewed','Sub_Categories_Viewed','Sub_Sub_Categories_Viewed','Products_Viewed']:
        counts = data[col].value_counts()
        zeros = counts.iloc[0]
        if zeros / len(data) * 100 >99:
            overfit.append(col)
print (len(overfit))
```

```python
# dropping overfitting columns
data.drop(overfit,axis=1,inplace=True)
```

```python
tmp = data['ProductList'].str.split(";",expand=True)
tmp =  tmp[0]
tmp = tmp.str.split("/",expand=True)
tmp.columns = ["Category","Sub_Category","Sub_Sub_Category","Product","Etc"]
tmp.drop("Etc",axis=1,inplace=True)
tmp
```

```python
new = pd.concat([data,tmp],axis=1)
new.head()
```

```python
category_frequency = new.Category.value_counts().to_dict()
new['Category_Frequency'] = new.Category.map(category_frequency)

sub_category_frequency = new.Sub_Category.value_counts().to_dict()
new['Sub_Category_Frequency'] = new.Sub_Category.map(sub_category_frequency)

sub_sub_category_frequency = new.Sub_Sub_Category.value_counts().to_dict()
new['Sub_Sub_Category_Frequency'] = new.Sub_Sub_Category.map(sub_sub_category_frequency)

product_frequency = new.Product.value_counts().to_dict()
new['Product_Frequency'] = new.Product.map(product_frequency)
```

```python
new ['gender'] = new.gender.apply(lambda x: 1 if x=="female" else 0)
```

```python
%%time
# Let's check the correlation between the variables and eliminate the one's that have high correlation
# Threshold for removing correlated variables
threshold = 0.9

# Absolute value correlation matrix
corr_matrix = new.corr().abs()
corr_matrix.head()
```

```python
# Upper triangle of correlations
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
upper.head()
```

```python
# Select columns with correlations above threshold
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

print('There are %d columns to remove.' % (len(to_drop)))
print ("Following columns can be dropped {}".format(to_drop))
```

```python
new.drop(to_drop,axis=1,inplace=True)
```

```python
X = train[[col for col in train.columns if col!="gender"]]
y = train['gender']
```

```python
kf = StratifiedKFold(n_splits=5,random_state=42,shuffle=False)
```

```python
# Let's tune some hyperparameters
clf_1 = lgb.LGBMClassifier(boosting_type='gbdt',objective='binary',random_state=42,n_jobs=-1,eval_metric='accuracy')
params = {"max_depth":[3,4,5,6,7,-1],
          "learning_rate":[0.01,0.05,0.1,0.3],
          "subsample":[0.5,0.6,0.7,0.8,0.9],
          "colsample_bytree":[0.5,0.6,0.7,0.8,0.9],
          "reg_alpha":[0.5,1,2,5,10],
          "reg_lambda":[0.5,1,2,5,10],
          "num_leaves":[7,15,31,63,127],
          "n_estimators":list(range(50,501,50)),
          "min_data_in_leaf":[1,3,5,10,15,25]}
random_search_1 = RandomizedSearchCV(estimator=clf_1,param_distributions=params,cv=kf.split(X,y),scoring='accuracy')
random_search_1.fit(X,y)
```

```python
random_search_1.best_params_,random_search_1.best_estimator_,random_search_1.best_score_
```

```python
plt.figure(figsize=(12,6))
ser = pd.Series(random_search_1.best_estimator_.feature_importances_,X.columns).sort_values()
ser.plot(kind='bar')
```

```python
# Columns that have feature importance more than 0
lst = list(ser[ser>0].index)
X = X[lst]
test = test[lst]
```
